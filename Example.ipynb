Create and train a model with all default values

```python
from forecasting import torchLSTM as tl
ticker = 'AAPL'
predictor = tl.PredictionEngine(ticker) #Creates model
predictor.train_ticker() # Trains model
predictor.eval_ticker() # Plots and prints results
predictor.save_model() # Saves Results
```

Create and train an hourly crypto model.

```python
ticker = 'ETH-USD'
predictor = tl.PredictionEngine(ticker,interval='1h',period='2y') #Creates model
predictor.train_ticker() # Trains model
predictor.eval_ticker() # Plots and prints results
predictor.save_model() # Stores model
```

Load a model and get next price predictions

```python
ticker = 'ETH-USD'
predictor = tl.PredictionEngine(ticker,is_loading=True) #Creates moddel

pos,pred_now,pred_prev = predictor.predict_now() # Get's prediction for tomorrow
print("Price moving up? " + str(pos))
print("Next prediction: " + str(pred_now))
print("Previos Prediction: " + str(pred_prev))
```

Loop through top 10 stocks and train a model using the Relative Normalization

```python
top_stocks = ['AAPL', 'MSFT', 'AMZN', 'GOOGL', 'FB', 'CMCSA', 'JPM', 'HD', 'DIS', 'XOM']
for stock in top_stocks:
    predictor = tl.PredictionEngine(stock,training_set_coeff=1,period='max',normalizer_type='Relative') #Creates model
    predictor.train_ticker() # Trains model
    predictor.save_model() # Stores model
    
```

Loop through top 10 stocks and print a prediction.

```python
top_stocks = ['AAPL', 'MSFT', 'AMZN', 'GOOGL', 'FB', 'CMCSA', 'JPM', 'HD', 'DIS', 'XOM']
for stock in top_stocks:
    predictor = tl.PredictionEngine(stock, is_loading=True)
    pos,pred_now,pred_prev = predictor.predict_now() # Get's prediction for tomorrow
    print("Price moving up? " + str(pos))
    print("Next prediction: " + str(pred_now))
    print("Previos Prediction: " + str(pred_prev))
```
